```python
import csv
```

```python
#imports required data: dates and names
rows = []
with open('input1.csv', 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    
    for row in csvreader: 
        rows.append(row)

dates = []; names = []
for x in rows: 
    dates.append(x[0])
    names.append(x[1])
```

```python
#creates list of nams and dates
nameslist = ['Anvit', 'Ayoola', 'Brenda', 'Brian', 'Corrie', 'Dennis', 'Eduardo', 'Heejin', 'Jess', 'Kimberley', 'Levy', 
         'Lishu','Perpetua', 'Sofia', 'Valentina']
dateslist = []
x = 1
while x <= 30:  
    adate = '12/' + str(x)
    dateslist.append(adate)
    x+=1
#print(dateslist)
```

```python
#checks if the individual checked in on a day

newfile = 'present2.csv'
with open(newfile, 'w', newline = '') as csvfile:
    filewriter = csv.writer(csvfile, quoting=csv.QUOTE_MINIMAL)
    for x in dateslist: 
        newlist = []
        for y in range(len(dates)): 
            if dates[y] == x: 
                newlist.append(names[y])
        #print(newlist)
        truelist = []
        truelist.append(x)
        for z in nameslist: 
            if z in newlist: 
                truelist.append('Yes')
            else: 
                truelist.append('No')
        #print(x, truelist)
        filewriter.writerow(truelist)
```
